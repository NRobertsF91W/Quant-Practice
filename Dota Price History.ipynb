```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date
```

```python
item_history = pd.read_csv("Dota Item Price Histories.csv")
display(item_history)
```

```python
data_to_plot = []
for item in np.unique(item_history["Item"]):
    xvals = item_history[item_history["Item"] == item]["Date"].to_numpy()
    yvals = item_history[item_history["Item"] == item]["Price"].to_numpy()
    data_to_plot.append([xvals,yvals])
    
fig1 = plt.figure(figsize=(10,7))
ax1 = fig1.add_subplot(111)
for m,list in enumerate(data_to_plot):
    ax1.plot(list[0],list[1], label= np.unique(item_history["Item"])[m])

ax1.set_xlabel("Date")
ax1.set_ylabel("Price ($)")
plt.legend();
```
